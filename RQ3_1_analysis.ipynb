```python
#no_deleted = df_comprehensive[pd.isna(df_comprehensive['Establishment'])]


item_versions = []

for index, row in df.iterrows():
    owner = row["Owner"]
    model_name = row["model_name"]
    pipeline = row["pipeline"]
    
    try:
        version_match = re.search(r'[-_]v(\d+(\.\d+)?)$', model_name)
        if version_match:
            version = version_match.group(1)
        else:
            version = 'none'
    except KeyError:
        version = "none"
            
    item_versions.append("v"+version)

df['ver'] = item_versions
def categorize_version(version):
    if version.startswith('v'):
        num_str = version[1:]
        if '.' not in num_str:
            return 'major'
        else:
            dot_count = num_str.count('.')
            if dot_count == 1:
                return 'minor'
            elif dot_count == 2:
                return 'patches'
            elif dot_count == 3:
                return 'negligent'
            else:
                return 'invalid'
    else:
        return 'invalid'


    
df['version_group'] = df['ver'].apply(categorize_version) 
df
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

base_compare1 = s_major['base_changed']
base_compare2 = s_minor['base_changed']

# Contingency table for chi-squared test
contingency_table = pd.crosstab(base_compare1, base_compare2)

# Perform chi-squared test
chi2, pval, degrees_of_freedom, expected_table = chi2_contingency(contingency_table.values)  # Use commas to unpack

# Print results
print("Chi-squared statistic:", chi2)
print("p-value:", pval)

# Interpretation
if pval < 0.05:
    print("Reject null hypothesis: Base model changes differ between major and minor versions.")
else:
    print("Fail to reject null hypothesis: Base model changes might be similar between major and minor versions.")
```

```python
from scipy.stats import fisher_exact

# Assuming base_compare1 and base_compare2 are pandas Series objects
contingency_table = pd.crosstab(base_compare1, base_compare2)

# Perform Fisher's exact test
odds_ratio, p = fisher_exact(contingency_table)

# Print results
print("Odds ratio:", odds_ratio)
print("P-value:", p)

# Decide whether to reject the null hypothesis
alpha = 0.05  # significance level
if p < alpha:
    print("Reject null hypothesis: There is a significant difference between the distributions.")
else:
    print("Fail to reject null hypothesis: There is no significant difference between the distributions.")
```

```python
from statsmodels.stats.proportion import proportions_ztest

base_compare1 = s_major['base_changed']
base_compare2 = s_minor['base_changed']

# Assuming base_compare1 and base_compare2 are pandas Series objects
successes1 = base_compare1.sum()  # Count of 1s in base_compare1
failures1 = len(base_compare1) - successes1  # Count of 0s in base_compare1
successes2 = base_compare2.sum()  # Count of 1s in base_compare2
failures2 = len(base_compare2) - successes2  # Count of 0s in base_compare2

# Perform two-sample proportion test
stat, pval = proportions_ztest([successes1, successes2], [len(base_compare1), len(base_compare2)])

print("Base Model Comparison")
# Print results
print("Z-statistic:", stat)
print("P-value:", pval)

epsilon = 0.000001

# Calculate odds ratio
odds_ratio = (successes1 / failures1) / ((successes2+epsilon) / (failures2+epsilon))
print("Odds ratio:", odds_ratio)

# Print the calculated odds ratio


# Decide whether to reject the null hypothesis
alpha = 0.05  # significance level
if pval < alpha:
    print("Reject null hypothesis: There is a significant difference between the proportions.")
else:
    print("Fail to reject null hypothesis: There is no significant difference between the proportions.")
```

```python
import json
import requests

def download_config(url):
    response = requests.get(url)
    if response.status_code == 200:
        return json.loads(response.text)
    else:
        print(f"Failed to download config from {url}. Status code: {response.status_code}")
        return None

def compare_configs(config1, config2):
    diff = {}
    for key, value in config1.items():
        if key in config2:
            if config2[key] != value:
                diff[key] = (value, config2[key])
        else:
            diff[key] = (value, None)

    for key, value in config2.items():
        if key not in config1:
            diff[key] = (None, value)

    return diff

# URLs of the JSON configurations
url1 = "https://huggingface.co/albert/albert-xlarge-v2/raw/main/config.json"
url2 = "https://huggingface.co/albert/albert-xlarge-v1/raw/main/config.json"

# Download configurations
config1 = download_config(url1)
config2 = download_config(url2)

if config1 is not None and config2 is not None:
    # Compare configurations
    differences = compare_configs(config1, config2)

    # Print differences
    for key, values in differences.items():
        print(f"Key: {key}")
        print(f"Config 1 value: {values[0]}")
        print(f"Config 2 value: {values[1]}")
        print()
else:
    print("Failed to download one or both configurations.")
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load the data into a DataFrame
data = {
    "Versions": ["Major", "Minor"],
    "Configuration": [36, 5],
    "Model Architecture": [12, 8],
    "Documentation": [50, 20],
    "Performance": [21, 1],
    "Dataset": [16, 3],
    "Training Library": [8, 3],
    "Energy consumption": [2, 0],
    "Performance metrics": [5, 1],
    "Other Changes": [22, 25]
}

# Create DataFrame
df = pd.DataFrame(data)
df.set_index("Versions", inplace=True)

# Perform chi-square test for each metric
results = {}
for column in df.columns:
    # Create contingency table
    table = [df[column].tolist()]
    chi2, p, dof, expected = chi2_contingency(table, correction=False)
    results[column] = {'chi2': chi2, 'p-value': p, 'dof': dof, 'expected': expected}

# Print results
for category, result in results.items():
    print(f"Metric: {category}")
    print(f"Chi-square statistic: {result['chi2']:.4f}")
    print(f"p-value: {result['p-value']:.4f}")
    print(f"Degrees of freedom: {result['dof']}")
    print(f"Expected frequencies: {result['expected']}")
    print()
```
