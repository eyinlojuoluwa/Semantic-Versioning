```python
import pandas as pd
```

```python
data = "naming_convention_terms.csv"
df = pd.read_csv(data)
df
```

```python
df['identifier'] = df.apply(lambda row: 1 if 'identifier' in row['naming_conv'] else 0, axis=1)
df['language'] = df.apply(lambda row: 1 if 'language' in row['naming_conv'] else 0, axis=1)
df['base-model'] = df.apply(lambda row: 1 if 'base-model' in row['naming_conv'] else 0, axis=1)
df['ambiguous'] = df.apply(lambda row: 1 if 'ambiguous' in row['naming_conv'] else 0, axis=1)
df['creation-date'] = df.apply(lambda row: 1 if 'creation-date' in row['naming_conv'] else 0, axis=1)
df['dataset'] = df.apply(lambda row: 1 if 'dataset' in row['naming_conv'] else 0, axis=1)
df['description'] = df.apply(lambda row: 1 if 'description' in row['naming_conv'] else 0, axis=1)
df['nationality'] = df.apply(lambda row: 1 if 'nationality' in row['naming_conv'] else 0, axis=1)
df['size'] = df.apply(lambda row: 1 if 'size' in row['naming_conv'] else 0, axis=1)
df['task'] = df.apply(lambda row: 1 if 'task' in row['naming_conv'] else 0, axis=1)
df['training-mechanism'] = df.apply(lambda row: 1 if 'training-mechanism' in row['naming_conv'] else 0, axis=1)
df['variant'] = df.apply(lambda row: 1 if 'variant' in row['naming_conv'] else 0, axis=1)
df['version'] = df.apply(lambda row: 1 if 'version' in row['naming_conv'] else 0, axis=1)

df.rename(columns={'selected_model': 'model_name'}, inplace=True)
df
```

```python
gp = df.groupby('naming_conv')['model_name'].nunique()
gp
```
